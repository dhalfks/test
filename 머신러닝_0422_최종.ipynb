## Kmeans 파이썬으로 구현

```python
import random
import numpy as np

data = []

for i in range(100):
    data.append([random.randint(140,180), random.randint(40,80)])
    data.append([random.randint(160,200), random.randint(60,100)])

print(data)
```

```python
import matplotlib.pyplot as plt

for i in data:
  plt.plot(i[0],i[1],'ok')
plt.show()
```

```python
def dist(x1, x2):  #함수 정의
    result = (x1[0]-x2[0])**2 + pow((x1[1]-x2[1]),2)   # 두점 사이의 거리를 구한 공식  공식처럼 알기
    result = np.sqrt(result)
    return result
```

```python
# KMeans 함수 정의
def kmeans(data, k, max_iter=100):  #data에서 k개만큼 군집하기  max_iter: 최대 이동의 횟수
    # 초기 중심점 선택
    centroids = random.sample(data, k) #random 모듈의 기능 
    
    # 반복적으로 중심점 이동
    for i in range(max_iter):
        # 각 데이터 포인트에 대해 가장 가까운 중심점 찾기
        clusters = [[] for _ in range(k)]   # list 내포 => 값을 초기화 할때 사용 (빈 공간을 k개만큼 생성)
        for i in data:
            distances = [dist(c,i) for c in centroids]  # c는 중심점 2개 중 하나 i 데이터의 점 중 하나  
            cluster_idx = np.argmin(distances)   #argmin : 둘중 작은것 찾기   argmax : 둘 중 큰것 찾기
            clusters[cluster_idx].append(i)

        # 새로운 중심점 계산
        new_centroids = []
        for i in clusters:
            new_centroid = np.mean(i, axis=0)    #mean : 평균 구하기 => 평균으로 새로운 점 구하기
            new_centroids.append(new_centroid)
        new_centroids = np.array(new_centroids)
        
        if np.allclose(centroids, new_centroids):  # allclose : 중심점의 오차가 적은형태면 break;
            break
            
        centroids = new_centroids
    
    return clusters

result = kmeans(data, k=2)

for i in result[0]:
    plt.plot(i[0],i[1],'or')
for i in result[1]:
    plt.plot(i[0],i[1],'ob')
plt.show()
```

## 선형회귀

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#공부시간 X와 성적 Y의 리스트를 만듭니다.
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]  # 독립변수 로 나누기
y = [i[1] for i in data]  # 종속변수

#그래프로 나타내 봅니다.
plt.figure(figsize=(8,5))
plt.scatter(x, y)   # 선점도 그래프
plt.show()

#리스트로 되어 있는 x와 y값을 넘파이 배열로 바꾸어 줍니다.(인덱스를 주어 하나씩 불러와 계산이 가능해 지도록 하기 위함입니다.)
x_data = np.array(x)   # 행렬 연산을 편하게 하기 위해 numpy array 형태로 변경
y_data = np.array(y)

# 기울기 a와 절편 b의 값을 초기화 합니다.
a = 0
b = 0

#학습률을 정합니다.
lr = 0.03   # 일반 적인 비율  너무 크거나(정답이 지나가버릴수 있음) 너무 작으면 (오래걸림)

#몇 번 반복될지를 설정합니다.
epochs = 2001 

#경사 하강법을 시작합니다.
for i in range(epochs): # epoch 수 만큼 반복
    y_hat = a * x_data + b  #y를 구하는 식을 세웁니다
    error = y_data - y_hat  #오차를 구하는 식입니다.
    a_diff = -(2/len(x_data)) * sum(x_data * (error)) # 오차함수를 a로 미분한 값입니다. 
    b_diff = -(2/len(x_data)) * sum(error)  # 오차함수를 b로 미분한 값입니다. 
    a = a - lr * a_diff  # 학습률을 곱해 기존의 a값을 업데이트합니다.
    b = b - lr * b_diff  # 학습률을 곱해 기존의 b값을 업데이트합니다.
    if i % 100 == 0:    # 100번 반복될 때마다 현재의 a값, b값을 출력합니다.
        print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i, a, b))

# 앞서 구한 기울기와 절편을 이용해 그래프를 그려 봅니다.
y_pred = a * x_data + b
plt.scatter(x, y)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
plt.show()

```

## 사이킷런을 이용한 선형회귀 처리

```python
from sklearn import linear_model
import numpy as np
X = np.array([2,4,6,8])
y = np.array([81, 93, 91, 97])
regr = linear_model.LinearRegression()  #선형회귀모델 사용하는 함수
X = X.reshape(len(X), -1)
print("입력(피쳐)\n", X)
print("출력(타겟)\n", y)
regr.fit(X, y) # 학습
print('w =', regr.coef_.round(2),', b =', regr.intercept_.round(2))  # regr.coef : 기울기, regr.intercept_: 절편
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(X, y, color='blue', marker='D')
# w, b를 이용하여 선형 회귀 직선을 그리자
y_pred = regr.coef_ * X + regr.intercept_ 
plt.plot(X, y_pred, 'r:')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

m = 100    # 생성할 데이터의 갯수
# 평균값이 0이고 -4에서 4사이에 분포하는 랜덤 값 X
X = 8 * np.random.rand(m, 1) - 4  # 100행 1열 난수 생성
X.shape
```

```python
# x^2항의 계수가 0.5, x항의 계수가 2, 상수항의 계수가 1
y = 0.5 * X ** 2 + 2 * X + 1 + np.random.randn(m, 1)
```

```python
poly_features = PolynomialFeatures(degree = 2,include_bias = False)
X_poly = poly_features.fit_transform(X)
regr = LinearRegression()
regr.fit(X_poly, y)
print('다항 회귀 모델의 점수 =', regr.score(X_poly, y).round(3))
print('다항 회귀 모델의 계수 =', regr.coef_, '절편 =', regr.intercept_)
```

```python
domain = np.linspace(-4, 4, 50).reshape(-1, 1)
domain_2 = poly_features.fit_transform(domain)
plt.figure(figsize=(6,4))
y_predict = regr.predict(domain_2)
plt.scatter(X, y)
plt.plot(domain, y_predict, color='r', linewidth=4)
plt.show()
```

```python
# 3차 방정식 만들기 
# 2개의 변곡점을 가진 데이터를 생성하고 시각화하는 코드
import numpy as np
import matplotlib.pyplot as plt

m = 100
X = 8 * np.random.rand(m, 1) - 4
y = .5 * X ** 3 + .5 * X ** 2 + X + 3 + np.random.randn(m, 1)
plt.figure(figsize=(6,5))
plt.plot(X, y, "b^")
plt.show()
```

```python
# 다항 회귀 모델을 만들고 점수와 계수, 절편을 출력하는 코드
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly_features = PolynomialFeatures(degree = 3,\
                                   include_bias = False)
X_poly = poly_features.fit_transform(X)
regr = LinearRegression()
regr.fit(X_poly, y)

print('다항 회귀 모델의 점수 =', regr.score(X_poly, y).round(3))
print('다항 회귀 모델의 계수 =', regr.coef_)
print('절편 =', regr.intercept_)

# 다항 회귀 곡선을 생성하는 기능 
X_new = np.linspace(-4, 4, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = regr.predict(X_new_poly)

plt.figure(figsize=(6,6))
plt.plot(X, y, "b^", label='Data')
plt.plot(X_new, y_new, "r-", label="Prediction")
plt.legend()
plt.show()
```

```python
# 과대적합
# 과대적합

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

np.random.seed(7)
# 20개의 데이터를 생성하고 시각화 하는 기능
m = 20
X1 = 8 * np.random.rand(m, 1) - 4
y1 = .5 * X1 ** 3 + .5 * X1 ** 2 + X1 + 3 + np.random.randn(m, 1)

# 20개의 데이터를 학습용 데이터, 테스트 데이터로 구분하자
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1)

plt.figure(figsize=(6,5))
plt.xlim(-5, 5); plt.ylim(-28, 50) # x, y 범위를 제한함
plt.plot(X_train1, y_train1, "b^") # 학습용 데이터 시각화
plt.show()
# 다항 회귀 모델을 만들고 점수와 계수, 절편을 출력
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 다항 특성이 20개가 되도록 하자
poly_features = PolynomialFeatures(degree = 20,\
                                   include_bias = False)
X_train_pl1 = poly_features.fit_transform(X_train1)
regr = LinearRegression()
regr.fit(X_train_pl1, y_train1)

print('다항 회귀 모델의 점수 =', regr.score(X_train_pl1, y_train1).round(3))
print('다항 회귀 모델의 계수 =', regr.coef_)
print('절편 =', regr.intercept_)
```

```python
# 테스트 데이터를 다항 회귀 모델의 입력으로 주고 예측 점수를 구해보자
X_test_pl1 = poly_features.fit_transform(X_test1)

print('테스트 데이터에 대한 모델의 점수 =', regr.score(X_test_pl1, y_test1)) # 헉!!!
```

** 리지, 라소 회귀



```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

np.random.seed(20)
# 20개의 데이터를 생성하자
m = 20
X2 = 8 * np.random.rand(m, 1) - 4
y2 = .5 * X2 ** 3 + .5 * X2 ** 2 + X2 + 3 + np.random.randn(m, 1)

# 20개의 데이터를 학습용 데이터, 테스트 데이터로 구분하자
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2)

# 다항 특성이 20개가 되도록 하자
poly_features2 = PolynomialFeatures(degree = 20,\
                                   include_bias = False)
X_train_pl2 = poly_features2.fit_transform(X_train2)
X_test_pl2 = poly_features2.fit_transform(X_test2)

regr = LinearRegression()
regr.fit(X_train_pl2, y_train2)

print('훈련 데이터의 점수 =', regr.score(X_train_pl2, y_train2).round(3))
print('테스트 데이터의 점수 =', regr.score(X_test_pl2, y_test2).round(3))
```

```python
### 규제를 추가해 보자 ##
from sklearn.preprocessing import StandardScaler

ss = StandardScaler() # 규제가 있는 모델은 스케일러로 변환을 해야 함
ss.fit(X_train_pl2) # 반드시 fit(학습)은 train 데이터에 대해서 한 번만....
train_scaled2, test_scaled2 = ss.transform(X_train_pl2),ss.transform(X_test_pl2)
ss.fit(y_train2)
y_train_scaled2, y_test_scaled2 = ss.transform(y_train2), ss.transform(y_test2) 
print(y_train_scaled2.shape)
print(y_test_scaled2.shape)
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha = 10)
ridge.fit(train_scaled2, y_train_scaled2)
print('훈련 데이터의 점수 =', ridge.score(train_scaled2, y_train_scaled2))
print('테스트 데이터의 점수 =', ridge.score(test_scaled2, y_test_scaled2))
```

```python
import matplotlib.pyplot as plt

train_scores = []
test_scores = []
alpha_values = [0.001, 0.01, 1, 10, 100, 1000]
for a in alpha_values:
  ridge = Ridge(alpha = a)
  ridge.fit(train_scaled2, y_train_scaled2)
  train_scores.append(ridge.score(train_scaled2, y_train_scaled2))
  test_scores.append(ridge.score(test_scaled2, y_test_scaled2))

plt.plot(np.log10(alpha_values), train_scores, 'r-', label='Train score')
plt.plot(np.log10(alpha_values), test_scores, 'b:', label='Test score')
plt.xlabel('Alpha value : 10^x scaled')
plt.ylabel('Ridge Regression')
plt.legend(loc="lower left")
plt.show()
```
