```python
import random
import math

data = [] # 1:여자, 0: 남자

for i in range(100):
    data.append([random.randint(140,180), random.randint(40,80),1 ])
    data.append([random.randint(160,200), random.randint(60,100), 0])

print(data)

hei = int(input('키를 입력해주세요: '))
wei = int(input('몸무게를 입력해주세요: '))
k = int(input('k 개수를 입력해주세요 :'))
new = [hei, wei]

# 입력한 데이터의 가장 가까운 k개의 데이터를 비교하는 코드

def dist(x1, x2):
    result = (x1[0]-x2[0])**2 + pow((x1[1]-x2[1]),2)   # 두점 사이의 거리를 구한 공식  공식처럼 알기
    result = math.sqrt(result)
    return result
```

```python
temp =[]
for i in data :
    temp.append([dist(i,new),i[2]])
# print(temp)
```

```python
count = 0
for i in temp[:k]:
    if i[1]==1:
      count+= 1
if count > k/2 :
  print('여')
else:
  print('남')
```

## Kmeans 파이썬으로 구현

```python
import random
import numpy as np

data = []

for i in range(100):
    data.append([random.randint(140,180), random.randint(40,80)])
    data.append([random.randint(160,200), random.randint(60,100)])

print(data)
```

```python
import matplotlib.pyplot as plt

for i in data:
  plt.plot(i[0],i[1],'ok')
plt.show()
```

```python
def dist(x1, x2):  #함수 정의
    result = (x1[0]-x2[0])**2 + pow((x1[1]-x2[1]),2)   # 두점 사이의 거리를 구한 공식  공식처럼 알기
    result = np.sqrt(result)
    return result
```

```python
# KMeans 함수 정의
def kmeans(data, k, max_iter=100):  #data에서 k개만큼 군집하기  max_iter: 최대 이동의 횟수
    # 초기 중심점 선택
    centroids = random.sample(data, k) #random 모듈의 기능 
    
    # 반복적으로 중심점 이동
    for i in range(max_iter):
        # 각 데이터 포인트에 대해 가장 가까운 중심점 찾기
        clusters = [[] for _ in range(k)]   # list 내포 => 값을 초기화 할때 사용 (빈 공간을 k개만큼 생성)
        for i in data:
            distances = [dist(c,i) for c in centroids]  # c는 중심점 2개 중 하나 i 데이터의 점 중 하나  
            cluster_idx = np.argmin(distances)   #argmin : 둘중 작은것 찾기   argmax : 둘 중 큰것 찾기
            clusters[cluster_idx].append(i)

        # 새로운 중심점 계산
        new_centroids = []
        for i in clusters:
            new_centroid = np.mean(i, axis=0)    #mean : 평균 구하기 => 평균으로 새로운 점 구하기
            new_centroids.append(new_centroid)
        new_centroids = np.array(new_centroids)
        
        if np.allclose(centroids, new_centroids):  # allclose : 중심점의 오차가 적은형태면 break;
            break
            
        centroids = new_centroids
    
    return clusters

result = kmeans(data, k=2)

for i in result[0]:
    plt.plot(i[0],i[1],'or')
for i in result[1]:
    plt.plot(i[0],i[1],'ob')
plt.show()
```
